```python
import pandas as pd
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
# Data digitized from datasheet using [WebPlotDigitizer](https://apps.automeris.io/wpd/)
led1={"model":"KA-2810AQBS-G", "manufacturer": "Kingbright","spectrum": pd.read_csv("Kingbright_LED_KA-2810AQBS-G.csv",names=["wavelength","QE"])}
```

```python
syto9_ex={"name":"Syto9+DNA absorption", "spectrum":pd.read_csv("syto9_dna_absorption.csv", names=["wavelength","QE"])}
```

```python
syto9_em={"name":"Syto9+DNA emission",
       "spectrum":pd.read_csv("syto9-dna-emission.csv",names=["wavelength","QE"])}
```

```python
eyy={"name":"768 Egg Yolk Yellow", "spectrum":pd.read_csv("768 Egg Yolk Yellow.csv", names=["wavelength","QE"])}
```

```python
fig, axes = plt.subplots(1,1)
axes.plot(led1["spectrum"]["wavelength"], led1["spectrum"]["QE"], label=led1["model"])
axes.plot(syto9_ex["spectrum"]["wavelength"], syto9_ex["spectrum"]["QE"], label=syto9_ex["name"])
axes.plot(syto9_em["spectrum"]["wavelength"], syto9_em["spectrum"]["QE"], label=syto9_em["name"])
axes.plot(eyy["spectrum"]["wavelength"], eyy["spectrum"]["QE"], label=eyy["name"])
axes.legend();
```
